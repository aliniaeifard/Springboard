```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import yfinance as yf 

# --- 1. Data Preparation ---

# Replace with actual stock symbols
stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'JPM', 'JNJ', 'V', 'PG', 'XOM'] 

# Download historical data (replace with your preferred data source)
start_date = '2020-01-01'
end_date = '2023-12-31'
data = yf.download(stock_symbols, start=start_date, end=end_date)['Adj Close']

# Calculate daily returns
returns = data.pct_change().dropna()

# --- 2. Portfolio Optimization Functions ---

def portfolio_return(weights, returns):
    """
    Calculates the expected portfolio return.

    Args:
        weights: Portfolio weights (array-like).
        returns: Daily returns of assets (DataFrame).

    Returns:
        Expected portfolio return.
    """
    return np.sum(returns.mean() * weights) * 252  # Annualized

def portfolio_volatility(weights, cov_matrix):
    """
    Calculates the portfolio volatility.

    Args:
        weights: Portfolio weights (array-like).
        cov_matrix: Covariance matrix of asset returns.

    Returns:
        Portfolio volatility.
    """
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)  # Annualized

def min_variance(cov_matrix):
    """
    Finds the portfolio with minimum variance.

    Args:
        cov_matrix: Covariance matrix of asset returns.

    Returns:
        Optimal weights and minimum portfolio volatility.
    """
    num_assets = len(cov_matrix)
    args = (cov_matrix,)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0, 1) for _ in range(num_assets))
    result = minimize(portfolio_volatility, num_assets*[1./num_assets], args=args,
                      method='SLSQP', bounds=bounds, constraints=constraints)
    return result.x, result.fun

def efficient_return(returns, cov_matrix, target_return):
    """
    Finds the portfolio with the minimum volatility for a given target return.

    Args:
        returns: Daily returns of assets (DataFrame).
        cov_matrix: Covariance matrix of asset returns.
        target_return: Target annualized portfolio return.

    Returns:
        Optimal weights and portfolio volatility.
    """
    num_assets = len(cov_matrix)
    args = (cov_matrix,)
    constraints = ({'type': 'eq', 'fun': lambda x: portfolio_return(x, returns) - target_return},
                   {'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0, 1) for _ in range(num_assets))
    result = minimize(portfolio_volatility, num_assets*[1./num_assets], args=args,
                      method='SLSQP', bounds=bounds, constraints=constraints)
    return result.x, result.fun

# --- 3. Calculate Portfolio Statistics ---

# Calculate covariance matrix
cov_matrix = returns.cov()

# Calculate minimum variance portfolio
min_vol_weights, min_vol = min_variance(cov_matrix)

# Generate efficient frontier data
target_returns = np.linspace(portfolio_return(min_vol_weights, returns), 0.2, 50)  # Adjust target return range as needed
efficient_frontier = []
for target in target_returns:
    weights, vol = efficient_return(returns, cov_matrix, target)
    efficient_frontier.append((weights, vol))

# --- 4. Display Weights for Each Target Return ---

#for target, (weights, vol) in zip(target_returns, efficient_frontier):
#    print(f"Target Return: {target:.2f}, Volatility: {vol:.2f}")
#    for symbol, weight in zip(stock_symbols, weights):
#        print(f"  {symbol}: {weight:.2%}")
#    print()




results = []

for target, (weights, vol) in zip(target_returns, efficient_frontier):
    row = {"Target Return": target, "Volatility": vol}
    # Add the stock weights to the row
    for symbol, weight in zip(stock_symbols, weights):
        row[symbol] = weight
    # Append the row to the results list
    results.append(row)

# Convert the results list into a DataFrame
df_results = pd.DataFrame(results)

# Display the DataFrame









# --- 5. Plot the Efficient Frontier and Individual Stock Returns ---

efficient_frontier_volatilities = [vol for weights, vol in efficient_frontier]

# Calculate individual stock returns and volatilities
stock_returns = returns.mean() * 252
stock_volatilities = returns.std() * np.sqrt(252)

plt.figure(figsize=(10, 6))

# Plot individual stocks
plt.scatter(stock_volatilities, stock_returns, marker='o', color='blue', label='Individual Stocks')
for i, symbol in enumerate(stock_symbols):
    plt.annotate(symbol, (stock_volatilities[i], stock_returns[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Plot efficient frontier
plt.scatter(efficient_frontier_volatilities, target_returns, marker='o', color='green', label='Efficient Frontier')

# Plot minimum variance portfolio
plt.scatter(min_vol, portfolio_return(min_vol_weights, returns), marker='*', color='red', label='Minimum Variance Portfolio')

plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Expected Return')
plt.title('Efficient Frontier and Individual Stock Returns')
plt.legend()
plt.grid(True)
plt.show()

df_results
```

# How about we use AR(p) or XGBoost for anticipating the returns of stocks? 

```python
import numpy as np
import pandas as pd
import yfinance as yf
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

def xgboost_pred_stock_9_lag(ticker):
# Download stock data from Yahoo Finance
    stock_symbol = ticker
    start_date = '2018-01-01'
    end_date = '2024-12-10'
    data = yf.download(stock_symbol, start=start_date, end=end_date)
    
    # Feature Engineering: Add lag features
    data['Lag1'] = data['Close'].shift(4)  # Previous day's close price
    data['Lag2'] = data['Close'].shift(5)  # 2 days ago close price
    data['Lag3'] = data['Close'].shift(6)  # 3 days ago close price
    data['Lag4'] = data['Close'].shift(7)  # 4 days ago close price
    data['Lag5'] = data['Close'].shift(8)  # 5 days ago close price
    data['Lag6'] = data['Close'].shift(9)  # 6 days ago close price
    data['Lag7'] = data['Close'].shift(10)  # 7 days ago close price
    data['Lag8'] = data['Close'].shift(11)  # 8 days ago close price
    data['Lag9'] = data['Close'].shift(12)  # 9 days ago close price

    # Remove rows with missing values due to lag
    data = data.dropna()
    
    # Target: Predict next day's close price
    data['Target'] = data['Close']
    data = data.dropna()
    
    # Features (X) and target (y)
    X = data[['Lag1', 'Lag2', 'Lag3','Lag4', 'Lag5', 'Lag6','Lag7', 'Lag8', 'Lag9' ]]
    y = data['Target']
    
    # Split the data into training and testing sets
    #X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.009, shuffle=False)
    
    #train_size = int(len(data) * 0.997)
    #X_train, X_test = X[:train_size], X[train_size:]
    #y_train, y_test = y[:train_size], y[train_size:]
    X_train, X_test = X[:int(len(data)-4)], X[int(len(data)-4):]
    y_train, y_test = y[:int(len(data)-4)], y[int(len(data)-4):]

    # Train an XGBoost regressor
    model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Plot the actual stock price vs predicted stock price
    plt.figure(figsize=(14, 7))
    
    plt.plot(data['Close'].tail(100), label='AAPL Adjusted Close Price')
    plt.title('Apple Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    # Plot actual stock price (y_test)
    plt.plot(data.index[-len(y_test):], y_test, label="Actual Stock Price", color='blue')
    
    # Plot predicted stock price (y_pred)
    plt.plot(data.index[-len(y_test):], y_pred, label="Predicted Stock Price", color='red', linestyle='--')

    # Add labels and title
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')
    plt.title('Stock Price Prediction vs Actual Stock Price')
    
    # Add a legend
    plt.legend()
    
    # Display the plot
    plt.grid(True)
    plt.show()



    # Calculate the Mean Squared Error
    mse = mean_squared_error(y_test, y_pred)
    return print(f'Mean Squared Error: {mse:.4f}')
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg

# Fetch historical stock price data for Apple
ticker = 'AAPL'
data = yf.download(ticker, start='2018-01-01', end='2024-12-10')

# Extract the adjusted closing prices
prices = data['Adj Close']

# Plot the historical stock prices
plt.figure(figsize=(10, 6))
plt.plot(prices, label='AAPL Adjusted Close Price')
plt.title('Apple Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Prepare the data for the AR model
train_size = int(len(prices) * 0.8)
train, test = prices[:train_size], prices[train_size:]

# Fit the AR(9) model
model = AutoReg(train, lags=9)
model_fitted = model.fit()

# Make predictions
predictions = model_fitted.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

# Plot the predictions
plt.figure(figsize=(10, 6))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test', color='orange')
plt.plot(test.index, predictions, label='Predicted', color='red')
plt.title('Apple Stock Price Prediction using AR(9) Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Calculate and print the RMSE
rmse = np.sqrt(np.mean((predictions - test) ** 2))
print(f'Root Mean Squared Error: {rmse:.2f}')
```
